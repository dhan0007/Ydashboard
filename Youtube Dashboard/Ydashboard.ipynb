```python
import dash
from dash import html, dcc

import dash_bootstrap_components as dbc
import plotly.express as px
import googleapiclient.discovery
from dash_extensions import Lottie
from dash.dependencies import Output, Input

API_KEY = 'AIzaSyBuW6X5Mh9Z26AAOKKySj00BfMdOsdLM0g'
CHANNEL_ID = 'JA-NQ4MtcRIog66wziD8fA'
youtube = googleapiclient.discovery.build('youtube', 'v3', developerKey=API_KEY)

url_subscribers = "https://lottie.host/c108af53-3c56-46af-8d81-2516abde5c33/LdITdGjp6S.json"
url_likes = "https://lottie.host/6402dd78-040d-49d6-9f76-0e5dff0dba00/IG8qJssSsl.json"
url_dislikes = "https://lottie.host/0be36be3-9262-4e9f-b4af-46acab11f1d9/7Giq4VNEny.json"
url_views = "https://lottie.host/2d2710e7-ab79-45c8-9a0f-fd0646146625/okqh4CqJFj.json"
url_shares = "https://lottie.host/99339c17-a2c0-41f4-bc33-8cdbbfedcafe/eNRpVjDLXu.json"
url_youtube = "https://lottie.host/6c310d32-f401-413e-945b-9135780674b2/XoaES70jyd.json"

img_src = "https://yt3.googleusercontent.com/ytc/APkrFKYtesP4MhpfHBo0-v7pZihkMT8FD6HcMAaF4Rzb7A=s176-c-k-c0x00ffffff-no-rj-mo"

options = dict(loop=True, autoplay=True, rendererSettings=dict(preserveAspectRatio='xMidYMid slice'))

# Function to get monthly and yearly views
def get_views_data():
    # Monthly views data
    request_monthly = youtube.youtubeAnalytics().reports().query(
        ids=f'channel=={CHANNEL_ID}',
        startDate='2022-01-01',
        endDate='2022-12-31',
        metrics='views',
        dimensions='month',
    )
    response_monthly = request_monthly.execute()

    # Yearly views data
    request_yearly = youtube.youtubeAnalytics().reports().query(
        ids=f'channel=={CHANNEL_ID}',
        startDate='2022-01-01',
        endDate='2022-12-31',
        metrics='views',
        dimensions='year',
    )
    response_yearly = request_yearly.execute()

    # Extract monthly and yearly views data
    monthly_views_data = response_monthly.get('rows', [])
    yearly_views_data = response_yearly.get('rows', [])

    return monthly_views_data, yearly_views_data







# Dash app
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.LUX])

# Layout of the app
app.layout = dbc.Container([
    dbc.Row([
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(Lottie(options=options, width="67%", height="67%", url=url_youtube)),
                dbc.CardBody([
                    dbc.Card([
                        dbc.CardImg(src=img_src),
                    ]),
                    dbc.Card([
                        dbc.CardBody([
                            dbc.CardLink("Sri Lanka Cricket", target="_blank", href="https://www.youtube.com/@srilankacricket")
                        ])
                    ])
                ])
            ]),
        ], width=2),
    ]),

    dbc.Row([
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(Lottie(options=options, width="25%", height="25%", url=url_subscribers)),
                dbc.CardBody([
                    html.H6('Subscribers'),
                    html.H2(id='content-reactions-1', children="000")
                ], style={'textAlign': 'center'})
            ]),
        ], width=2),
    ]),

    dbc.Row([
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(Lottie(options=options, width="67%", height="67%", url=url_likes)),
                dbc.CardBody([
                    html.H6('Likes'),
                    html.H2(id='content-connections', children="000")
                ], style={'textAlign': 'center'})
            ]),
        ], width=2),
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(Lottie(options=options, width="32%", height="32%", url=url_dislikes)),
                dbc.CardBody([
                    html.H6('Dislikes'),
                    html.H2(id='content-companies', children="000")
                ], style={'textAlign': 'center'})
            ]),
        ], width=2),
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(Lottie(options=options, width="25%", height="25%", url=url_shares)),
                dbc.CardBody([
                    html.H6('Shares'),
                    html.H2(id='content-msg-in', children="000")
                ], style={'textAlign': 'center'})
            ]),
        ], width=2),
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(Lottie(options=options, width="53%", height="53%", url=url_views)),
                dbc.CardBody([
                    html.H6('Views'),
                    html.H2(id='content-msg-out', children="000")
                ], style={'textAlign': 'center'})
            ]),
        ], width=2),
    ]),

    dbc.Row([
        dbc.Col([
            dbc.Card([
                dbc.CardBody([
                    html.H4("Total Videos", className="card-title"),
                    html.P(id="total_videos", className="card-text"),
                ])
            ]),
        ], width=4),
        dbc.Col([
            dbc.Card([
                dbc.CardBody([
                    html.H4("Total Views", className="card-title"),
                    html.P(id="total_views", className="card-text"),
                ]),
            ]),
        ], width=4),
    ], className="mb-4"),

    dbc.Row([
        dbc.Col([
            dbc.Card([
                dbc.CardBody([
                    html.H4("Total Watch Time", className="card-title"),
                    html.P(id="total_watch_time", className="card-text"),
                ])
            ]),
        ], width=4),
        dbc.Col([
            dbc.Card([
                dbc.CardBody([
                    html.H4("Monthly Views Chart", className="card-title"),
                    dcc.Graph(id='monthly_views_chart'),
                ])
            ]),
        ], width=8),
    ]),

    dbc.Row([
        dbc.Col([
            dbc.Card([
                dbc.CardBody([
                    html.H4("Yearly Views Chart", className="card-title"),
                    dcc.Graph(id='yearly_views_chart'),
                ])
            ]),
        ], width=12),
    ]),

    
    html.Div(id='trending_video_details'),

], fluid=True)

# Update the callback input component id
@app.callback(
    [dash.dependencies.Output("monthly_views_chart", "figure"),
     dash.dependencies.Output("yearly_views_chart", "figure"),
     dash.dependencies.Output("trending_video_details", "children")],
    [dash.dependencies.Input("monthly_views_chart", "figure")])
def update_charts_and_card(monthly_views_chart_figure):
    # Get monthly and yearly views data
    monthly_views_data, yearly_views_data = get_views_data()

    # Create monthly views chart
    monthly_views_chart = px.bar(monthly_views_data, x='month', y='views', title='Monthly Views')

    # Create yearly views chart
    yearly_views_chart = px.bar(yearly_views_data, x='year', y='views', title='Yearly Views')

    # Get details of the trending video 
    trending_video_details = f"Video Title: XYZ | Views: {monthly_views_data[-1]['views']}"

    return monthly_views_chart, yearly_views_chart, trending_video_details

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
